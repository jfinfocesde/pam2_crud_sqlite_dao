```python
import pandas as pd
import numpy as np

# Generar 50 filas con 5 columnas de datos aleatorios
data = {
    'Edad': np.random.randint(18, 65, 50),
    'Ingreso': np.random.uniform(20000, 100000, 50),
    'Género': np.random.choice(['Hombre', 'Mujer'], 50),
    'Ciudad': np.random.choice(['Madrid', 'Barcelona', 'Valencia'], 50),
    'Fecha_Nacimiento': pd.date_range('1960-01-01', periods=50)
}

# Crear un DataFrame de pandas
df = pd.DataFrame(data)

# Guardar el DataFrame como un archivo CSV
df.to_csv('datos_ejemplo.csv', index=False)
```
